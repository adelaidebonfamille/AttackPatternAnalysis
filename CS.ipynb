If the difference between the mean and the 75th percentile of 'Source Port' and 'Destination Port' is very different, but the minimum and maximum values are the same, then we can use hypothesis testing to determine whether the difference is statistically significant.


$$ H_0: \mu_1=\mu_2$$
$$ H_a: \mu_1\neq\mu_2$$

The appropriate hypothesis test for this case is the T-test. The T-test compares the means between two groups of data. In this case, we will compare the mean of 'Source Port' and 'Destination Port'.

Null hypothesis ($H_0$): There is no significant difference between the means of 'Source Port' and 'Destination Port'.
Alternative hypothesis ($H_1$): There is a significant difference between the means of 'Source Port' and 'Destination Port'.

We will then use the p-value from the T-test to determine whether we should reject or retain the null hypothesis. If the p-value is less than the predetermined significance level ($\alpha$), we will reject the null hypothesis and conclude that the difference between the means of 'Source Port' and 'Destination Port' is statistically significant. If the p-value is greater than $\alpha$, we will retain the null hypothesis and conclude that there is no significant difference between the means of the two variables.

```python
statistic, pvalue = stats.ttest_ind( newdf['Source Port'], newdf['Destination Port'], equal_var=False)
print('p-value in T-test: ' + str(pvalue))
```

Due to the $p$-value being extremely small, Python approximates it to 0.0. This suggests that we can confidently reject the null hypothesis $H_0$ which posits the equality of means between the source and destination ports. Therefore, the source port of attacks significantly differs from the destination port. Consequently, it indicates that most attacks are targeted towards specific logical ports rather than occurring randomly.

```python
def heatmap_graph(df, xlabel, ylabel, title):
    plt.figure(figsize=(18,8))
    ax = sns.heatmap(df)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()
```

The machine with the IPv4 address 149.171.126.17 is the primary target of attacks. On the other hand, while worms, shellcode, and generic attacks are indiscriminate and not aimed at specific machines, Denial of Service, Exploits, and Backdoor attacks clearly focus on specific servers.

```python
for attack in list(newdf['Attack category'].unique()):
    df_attack = newdf[newdf['Attack category'] == attack].copy()
    statistic, pvalue = stats.ttest_ind(df_attack['Source Port'], df_attack['Destination Port'], equal_var=False)
    print('p-value in T-test for ' + attack + ' attack: ' + str(pvalue))
```

As observed, the p-values for all attack categories, except for Shellcode attacks, are extremely close to 0.0. This suggests that the attacks have been targeted towards specific ports. However, for Shellcode attacks, the null hypothesis cannot be rejected. This implies that there is a certain level of randomness associated with this type of attack, indicating that the average values of source and destination ports are similar.

To validate this assertion, we will utilize a contingency table, enabling us to examine the frequency count of specific pairs of variables, similar to the approach demonstrated with .pivot_table().

```python
df_crosstab = pd.crosstab(newdf['Attack category'], newdf['Destination Port'])
df_crosstab
```

```python
chi2, p_value, dof, expected = chi2_contingency(df_crosstab)
print("p-value of Chi-square test for Attack category vs. Destination Port =", p_value)
```
